# Base Load

```python
import os.path
import time
import random
import csv
import array
import math
import numpy as np
import pandas as pd
import scipy
import scipy.io
from joblib import Parallel, delayed
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from pypower.api import ext2int, int2ext, makeYbus

class BaseLoad:
    def __init__(self, path, test_case, optimization_method='TNC'):
        self.PATH = path
        self.TEST_CASE = test_case
        self.OPTIMIZATION_METHOD = optimization_method
        self.baseMVA = None
        self.bus = None
        self.gen = None
        self.branch = None
        self.Ybus = None
        self.Yf = None
        self.Yt = None
        self.ppc = None
        self.load_buses = None
        self.n_buses = None
        self.n_gens = None
        self.n_branches = None
        self.r = None
        self.x = None
        self.ref_bus = None
        self.F = None
        self.V_AC = None
        self.theta_AC = None

    def load_mat(self):
        mat = scipy.io.loadmat(f"{self.PATH}/data/case{self.TEST_CASE}.mat")
        self.bus = mat['ans']['bus'][0][0]
        self.baseMVA = mat['ans']['baseMVA'][0][0][0][0]
        self.gen = mat['ans']['gen'][0][0]
        self.branch = mat['ans']['branch'][0][0]
        keys = ['baseMVA', 'bus', 'gen', 'branch']
        values = [self.baseMVA, self.bus, self.gen, self.branch]
        self.ppc = dict(zip(keys, values))
        self.ppc = ext2int(self.ppc)
        self.baseMVA, self.bus, self.gen, self.branch = self.ppc['baseMVA'], self.ppc['bus'], self.ppc['gen'], self.ppc['branch']
        self.Ybus, self.Yf, self.Yt = makeYbus(self.baseMVA, self.bus, self.branch)
        return self.baseMVA, self.bus, self.gen, self.branch, self.Ybus, self.Yf, self.Yt, self.ppc

    def process_bus_and_gen(self):
        Just_load = self.bus[np.logical_or(self.bus[:, 2] != 0, self.bus[:, 3] != 0)]
        self.load_buses = Just_load[:, 0]
        self.load_buses = np.reshape(self.load_buses, (len(self.load_buses), 1))
        self.n_buses = len(self.ppc['bus'])
        self.n_gens = len(self.ppc['gen'])
        self.n_branches = len(self.ppc['branch'])
        self.r = self.branch[:, 2]
        self.x = self.branch[:, 3]
        self.ref_bus = np.where(self.bus[:, 1] == 3)[0][0]
        return self.load_buses, self.n_buses, self.n_gens, self.n_branches, self.r, self.x, self.ref_bus

    def compute_incidence_matrix(self):
        A = np.zeros((self.n_buses, self.n_branches))
        from_bus = self.branch[:, 0].astype(int)
        to_bus = self.branch[:, 1].astype(int)
        A[from_bus, np.arange(self.n_branches)] = 1
        A[to_bus, np.arange(self.n_branches)] = -1
        A = np.delete(A, self.ref_bus, axis=0)
        A = A.T
        return A

    def load_data(self):
        #AC_DATA   = pd.read_csv (f"{self.PATH}/results/High_K=2_Voltage_AC_{self.TEST_CASE}bus.txt", sep=',')
        AC_DATA = pd.read_csv(f"{self.PATH}/results/Base_Voltage_AC_{self.TEST_CASE}bus.txt", sep=',')
        df = pd.DataFrame(AC_DATA, columns=['i', 'Vm', 'Va'])
        self.V_AC = df.Vm
        self.theta_AC = df.Va
        return self.V_AC, self.theta_AC

    def LinDistFlow(self, Dr, Dx, gamma_P, gamma_Q, bias, ref_bus, n_buses, n_branches, F, P_inj, Q_inj, V_AC):
        #global gradient,V_AC, gradient_bias

        Dr = np.reshape(Dr, (n_branches, 1))
        Dx = np.reshape(Dx, (n_branches, 1))
        gamma_P = np.reshape(gamma_P, (n_buses - 1, 1))
        gamma_Q = np.reshape(gamma_Q, (n_buses - 1, 1))
        bias = np.reshape(bias, (n_buses - 1, 1))

        I = np.reshape(np.ones(n_buses - 1), (n_buses - 1, 1))
        I = 1.05 * I
        v = I + 2 * F @ np.multiply(Dr, F.T) @ (P_inj + gamma_P) + 2 * F @ np.multiply(Dx, F.T) @ (Q_inj + gamma_Q) + bias

        v_AC = np.reshape(V_AC.values, (n_buses, 1))
        v_AC = np.delete(v_AC, ref_bus, axis=0)

        v_AC = np.reshape(v_AC, (n_buses - 1, 1))
        v = np.reshape(v, (n_buses - 1, 1))

        return v, v_AC

    def objective_function(self, Dr, Dx, gamma_P, gamma_Q, bias, F, ref_bus, n_buses, n_branches, load_buses, P_inj_all_scenarios, Q_inj_all_scenarios, V_AC):
        #global  v, v_AC
        V_LDF = []
        V_DF = []

        v_results, v_AC_results = self.LinDistFlow(Dr, Dx, gamma_P, gamma_Q, bias, ref_bus, n_buses, n_branches, F, P_inj_all_scenarios, Q_inj_all_scenarios, V_AC)

        V_LDF = np.concatenate(v_results)
        V_DF = np.concatenate(v_AC_results)

        #objective = (1/(n_buses-1))*(1/Batch_size)*np.sum((V_LDF-V_DF)**2)
        objective = (1 / (n_buses - 1)) * np.linalg.norm(V_LDF ** 0.5 - V_DF, 1)
        inf_norm_error = np.linalg.norm(V_LDF ** 0.5 - V_DF, np.inf)

        return objective, inf_norm_error, V_LDF, V_DF

    def convert_branch(self):
        Vbase = self.bus[0, 9] * 1e3  # Vbase = mpc.bus(1, BASE_KV) * 1e3; in Volts
        Sbase = self.baseMVA * 1e6  # Sbase = mpc.baseMVA * 1e6; in VA
        self.branch[:, [2, 3]] = self.branch[:, [2, 3]] / (Vbase ** 2 / Sbase)  # convert branch impedance

    def run_analysis(self):
        baseMVA, bus, gen, branch, Ybus, Yf, Yt, ppc = self.load_mat()

        load_buses, n_buses, n_gens, n_branches, r, x, ref_bus = self.process_bus_and_gen()
        V_AC, theta_AC = self.load_data()
        A = self.compute_incidence_matrix()
        F = np.linalg.inv(A)

        """ For distribution systems that are not in per unit (22,33,69,85) cases"""
        #P_inj_all_scenarios= - bus[1:,2]/(1000*baseMVA)
        #Q_inj_all_scenarios= - bus[1:,3]/(1000*baseMVA)

        """ For distribution systems that are in per unit (123) """
        P_inj_all_scenarios = -np.delete(bus[:, 2], ref_bus) / baseMVA
        Q_inj_all_scenarios = -np.delete(bus[:, 3], ref_bus) / baseMVA

        """Only For 141 bus case which is different and only has MVA data"""
        #pf = 0.85
        #bus[:, 3] = bus[:, 2] * np.sin(np.arccos(pf))  # MVA to MVAr
        #bus[:, 2] = bus[:, 2] * pf  # MVA to MW
        #P_inj_all_scenarios= -np.delete(bus[:,2], ref_bus)/(1000*baseMVA)
        #Q_inj_all_scenarios= -np.delete(bus[:,3], ref_bus)/(1000*baseMVA)
        """Make sure to comment above for the other case"""

        P_inj_all_scenarios = np.reshape(P_inj_all_scenarios, (n_buses - 1, 1))
        Q_inj_all_scenarios = np.reshape(Q_inj_all_scenarios, (n_buses - 1, 1))

        """For distribution systems that are not in per unit(22,33,69,85,141) cases"""
        #self.convert_branch(baseMVA, bus, branch)
        r = branch[:, 2]
        x = branch[:, 3]

        ## Cold-start parameters
        Dr = r
        Dx = x
        gamma_P = np.zeros(n_buses - 1)
        gamma_Q = np.zeros(n_buses - 1)
        bias = np.zeros(n_buses - 1)

        # Optimized parameters
        #Dr = np.loadtxt(f"{self.PATH}/parameters/Dr_{self.TEST_CASE}bus_{self.OPTIMIZATION_METHOD}.txt")
        #Dx = np.loadtxt(f"{self.PATH}/parameters/Dx_{self.TEST_CASE}bus_{self.OPTIMIZATION_METHOD}.txt")
        #gamma_P = np.loadtxt(f"{self.PATH}/parameters/gamma_P_{self.TEST_CASE}bus_{self.OPTIMIZATION_METHOD}.txt")
        #gamma_Q = np.loadtxt(f"{self.PATH}/parameters/gamma_Q_{self.TEST_CASE}bus_{self.OPTIMIZATION_METHOD}.txt")
        #bias = np.loadtxt(f"{self.PATH}/parameters/bias_{self.TEST_CASE}bus_{self.OPTIMIZATION_METHOD}.txt")

        Dr = np.reshape(Dr, (n_branches,))
        Dx = np.reshape(Dx, (n_branches,))
        gamma_P = np.reshape(gamma_P, (n_buses - 1,))
        gamma_Q = np.reshape(gamma_Q, (n_buses - 1,))
        bias = np.reshape(bias, (n_buses - 1,))

        start_time = time.process_time()
        objective, inf_norm_error, V_LDF, V_DF = self.objective_function(Dr, Dx, gamma_P, gamma_Q, bias, F, ref_bus, n_buses, n_branches, load_buses, P_inj_all_scenarios, Q_inj_all_scenarios, V_AC)
        end_time = time.process_time()
        execution_time = end_time - start_time

        print(f'Max Error: {inf_norm_error}')
        print(f'Avg error {self.TEST_CASE} Bus: ', objective)
        print(f'Total execution time: {execution_time} seconds')

        np.savetxt(f"{self.PATH}/parameters/V_LinDistFlow_{self.TEST_CASE}_base.txt", V_LDF)

if __name__ == "__main__":
    TEST_CASE           =   '906'
    OPTIMIZATION_METHOD =   'TNC'
    PATH = '/Users/babaktaheri/Desktop/OLDF/Single-phase' 

    analysis = BaseLoad(PATH, TEST_CASE, OPTIMIZATION_METHOD)
    analysis.run_analysis()
```

```python
import os.path
import time
import random
import csv
import array
import math
import numpy as np
import pandas as pd
import scipy
import scipy.io
from joblib import Parallel, delayed
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from pypower.api import ext2int, int2ext, makeYbus

class PowerSystemAnalysis:
    def __init__(self, path, test_case, optimization_method='TNC'):
        self.PATH = path
        self.TEST_CASE = test_case
        self.OPTIMIZATION_METHOD = optimization_method
        self.baseMVA = None
        self.bus = None
        self.gen = None
        self.branch = None
        self.Ybus = None
        self.Yf = None
        self.Yt = None
        self.ppc = None
        self.load_buses = None
        self.n_buses = None
        self.n_gens = None
        self.n_branches = None
        self.r = None
        self.x = None
        self.ref_bus = None
        self.F = None
        self.V_AC = None
        self.theta_AC = None

    def load_mat(self):
        mat = scipy.io.loadmat(f"{self.PATH}/data/case{self.TEST_CASE}.mat")
        self.bus = mat['ans']['bus'][0][0]
        self.baseMVA = mat['ans']['baseMVA'][0][0][0][0]
        self.gen = mat['ans']['gen'][0][0]
        self.branch = mat['ans']['branch'][0][0]
        keys = ['baseMVA', 'bus', 'gen', 'branch']
        values = [self.baseMVA, self.bus, self.gen, self.branch]
        self.ppc = dict(zip(keys, values))
        self.ppc = ext2int(self.ppc)
        self.baseMVA, self.bus, self.gen, self.branch = self.ppc['baseMVA'], self.ppc['bus'], self.ppc['gen'], self.ppc['branch']
        self.Ybus, self.Yf, self.Yt = makeYbus(self.baseMVA, self.bus, self.branch)
        return self.baseMVA, self.bus, self.gen, self.branch, self.Ybus, self.Yf, self.Yt, self.ppc

    def process_bus_and_gen(self):
        Just_load = self.bus[np.logical_or(self.bus[:, 2] != 0, self.bus[:, 3] != 0)]
        self.load_buses = Just_load[:, 0]
        self.load_buses = np.reshape(self.load_buses, (len(self.load_buses), 1))
        self.n_buses = len(self.ppc['bus'])
        self.n_gens = len(self.ppc['gen'])
        self.n_branches = len(self.ppc['branch'])
        self.r = self.branch[:, 2]
        self.x = self.branch[:, 3]
        self.ref_bus = np.where(self.bus[:, 1] == 3)[0][0]
        return self.load_buses, self.n_buses, self.n_gens, self.n_branches, self.r, self.x, self.ref_bus

    def compute_incidence_matrix(self):
        A = np.zeros((self.n_buses, self.n_branches))
        from_bus = self.branch[:, 0].astype(int)
        to_bus = self.branch[:, 1].astype(int)
        A[from_bus, np.arange(self.n_branches)] = 1
        A[to_bus, np.arange(self.n_branches)] = -1
        A = np.delete(A, self.ref_bus, axis=0)
        A = A.T
        return A

    def load_data(self):
        AC_DATA = pd.read_csv(f"{self.PATH}/results/High_Voltage_AC_{self.TEST_CASE}bus.txt", sep=',')
        df = pd.DataFrame(AC_DATA, columns=['i', 'Vm', 'Va'])
        self.V_AC = df.Vm
        self.theta_AC = df.Va
        return self.V_AC, self.theta_AC

    def LinDistFlow(self, Dr, Dx, gamma_P, gamma_Q, bias, s, ref_bus, n_buses, n_branches, F, P_inj, Q_inj, V_AC):
        Dr = np.reshape(Dr, (n_branches, 1))
        Dx = np.reshape(Dx, (n_branches, 1))
        gamma_P = np.reshape(gamma_P, (n_buses - 1, 1))
        gamma_Q = np.reshape(gamma_Q, (n_buses - 1, 1))
        bias = np.reshape(bias, (n_buses - 1, 1))

        I = np.reshape(np.ones(n_buses - 1), (n_buses - 1, 1))
        I = 1.05 * I
        v = I + 2 * F @ np.multiply(Dr, F.T) @ (P_inj + gamma_P) + 2 * F @ np.multiply(Dx, F.T) @ (Q_inj + gamma_Q) + bias

        v_AC = V_AC[n_buses * s:n_buses * (s + 1)]
        v_AC = np.array(v_AC)
        v_AC = np.reshape(v_AC, (n_buses, 1))
        v_AC = np.delete(v_AC, ref_bus, axis=0)

        v_AC = np.reshape(v_AC, (n_buses - 1, 1))
        v = np.reshape(v, (n_buses - 1, 1))

        return v, v_AC

    def objective_function(self, Dr, Dx, gamma_P, gamma_Q, bias, F, ref_bus, n_buses, n_branches, load_buses, Batch_size, scen_start, scen_final, P_inj_all_scenarios, Q_inj_all_scenarios, V_AC):
        V_LDF = []
        V_DF = []

        v_results, v_AC_results = zip(*Parallel(n_jobs=-1)(delayed(self.LinDistFlow)(Dr, Dx, gamma_P, gamma_Q, bias, s, ref_bus, n_buses, n_branches, F, P_inj_all_scenarios[s], Q_inj_all_scenarios[s], V_AC) for s in range(scen_start - 1, scen_final - 1)))
        V_LDF = np.concatenate(v_results)
        V_DF = np.concatenate(v_AC_results)

        # objective = (1/(n_buses-1))*(1/Batch_size)*np.sum((V_LDF-V_DF)**2)
        objective = (1 / (n_buses - 1)) * (1 / Batch_size) * np.linalg.norm(V_LDF ** 0.5 - V_DF, 1)
        inf_norm_error = np.linalg.norm(V_LDF ** 0.5 - V_DF, np.inf)

        return objective, inf_norm_error, V_LDF, V_DF

    def convert_branch(self):
        Vbase = self.bus[0, 9] * 1e3
        Sbase = self.baseMVA * 1e6
        self.branch[:, [2, 3]] = self.branch[:, [2, 3]] / (Vbase ** 2 / Sbase)

    def run(self):
        start_time = time.time()
        self.load_mat()
        self.process_bus_and_gen()
        self.load_data()
        A = self.compute_incidence_matrix()
        self.F = np.linalg.inv(A)

        scen_start = 1
        scen_final = 31
        Batch_size = scen_final - scen_start
        P_inj_all_scenarios = np.load(f"{self.PATH}/data/High_P_inj_all_scenarios_{self.TEST_CASE}bus.npy", allow_pickle=True)
        Q_inj_all_scenarios = np.load(f"{self.PATH}/data/High_Q_inj_all_scenarios_{self.TEST_CASE}bus.npy", allow_pickle=True)

        """ For distribution systems that are not in per unit"""
        # self.convert_branch()
        self.r = self.branch[:, 2]
        self.x = self.branch[:, 3]

        """ Cold-start parameters"""
        Dr = self.r
        Dx = self.x
        gamma_P = np.zeros(self.n_buses - 1)
        gamma_Q = np.zeros(self.n_buses - 1)
        bias = np.zeros(self.n_buses - 1)

        """ Optimized parameters"""
        #Dr = np.loadtxt(f"{self.PATH}/parameters/Dr_{self.TEST_CASE}bus_{self.OPTIMIZATION_METHOD}.txt")
        #Dx = np.loadtxt(f"{self.PATH}/parameters/Dx_{self.TEST_CASE}bus_{self.OPTIMIZATION_METHOD}.txt")
        #gamma_P = np.loadtxt(f"{self.PATH}/parameters/gamma_P_{self.TEST_CASE}bus_{self.OPTIMIZATION_METHOD}.txt")
        #gamma_Q = np.loadtxt(f"{self.PATH}/parameters/gamma_Q_{self.TEST_CASE}bus_{self.OPTIMIZATION_METHOD}.txt")
        #bias = np.loadtxt(f"{self.PATH}/parameters/bias_{self.TEST_CASE}bus_{self.OPTIMIZATION_METHOD}.txt")

        Dr = np.reshape(Dr, (self.n_branches,))
        Dx = np.reshape(Dx, (self.n_branches,))
        gamma_P = np.reshape(gamma_P, (self.n_buses - 1,))
        gamma_Q = np.reshape(gamma_Q, (self.n_buses - 1,))
        bias = np.reshape(bias, (self.n_buses - 1,))

        objective, inf_norm_error, V_LDF, V_DF = self.objective_function(Dr, Dx, gamma_P, gamma_Q, bias, self.F, self.ref_bus, self.n_buses, self.n_branches, self.load_buses, Batch_size, scen_start, scen_final, P_inj_all_scenarios, Q_inj_all_scenarios, self.V_AC)
        print(f'Max Norm Error: {inf_norm_error}')
        print(f'Avg error {self.TEST_CASE} Bus: ', objective)

        # np.savetxt(f"{self.PATH}/parameters/V_LinDistFlow_{self.TEST_CASE}_{self.OPTIMIZATION_METHOD}.txt", V_LDF)
        # np.savetxt(f"{self.PATH}/parameters/V_DistFlow_{self.TEST_CASE}.txt", V_DF)
        # np.savetxt(f"{self.PATH}/parameters/V_cold_{self.TEST_CASE}.txt", V_LDF)
        end_time = time.time()
        execution_time = end_time - start_time
        print(f'Total execution time: {execution_time} seconds')

if __name__ == "__main__":
    TEST_CASE           =   '906'
    OPTIMIZATION_METHOD =   'TNC'
    PATH = '/Users/babaktaheri/Desktop/OLDF/Single-phase' 

    analysis = PowerSystemAnalysis(PATH, TEST_CASE, OPTIMIZATION_METHOD)
    analysis.run()
```
